```python
import pandas as pd
import json
```

```python
target = 'test'
```

```python
f = open(f'{target}.json') 
```

```python
data = json.load(f)
```

```python
utts = []
labels = []
for dialog in data:
    for window in dialog:
        utts.append('\n'.join([i for i in window['utterances'] if i != '']))
        labels.append(window['label'])
```

```python
onto = json.load(open('ontology.json'))
```

```python
onto
```

```python
mapping_list = list()
for i in onto:
    if i != '状态':
        mapping_list += onto[i]
```

```python
mapping_dict = dict()
for idx,i in enumerate(mapping_list):
    mapping_dict[i] = idx
```

```python
mapping_dict
```

```python
map_dict= dict((v, k) for k, v in mapping_dict.items())
with open("mapping_dict.json", "w") as fp:
    json.dump(map_dict,fp) 
```

```python
generated_labels = []
for i in labels:
    generated_labels.append([mapping_dict[l.split('-')[0].split(':')[1]] for l in i])
```

```python
generated_labels
```

```python
df = pd.DataFrame({'content':utts, 'label':[i for i in generated_labels],'category':['症状；检查；手术；其他信息']*len(utts)})
```
